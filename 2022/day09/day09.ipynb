```python
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
head_moves = []

with open('input.txt','r') as f :
    for line in f:
        head_moves.append(line.replace('\n',''))

head_moves
```

```python
def move_tail(head_x: int, head_y: int, tail_x: int, tail_y: int):
    diff_x, diff_y = head_x - tail_x, head_y - tail_y
    match diff_x, diff_y:
        # diagonal moves
        case(1,2) | (2,1): return (tail_x+1, tail_y+1) # move right+up
        case(1,-2) | (2, -1): return (tail_x+1, tail_y-1) # move right+down
        case(-1,2) | (-2,1): return (tail_x-1, tail_y+1) # move left+up
        case(-1,-2) | (-2,-1): return (tail_x-1, tail_y-1) # move left+down

        #horizontal moves
        case(2,0): return (tail_x+1, tail_y) # move right
        case(-2,0): return (tail_x-1, tail_y) # move left

        #vertical moves
        case(0,2): return (tail_x, tail_y+1) # move up
        case(0,-2): return (tail_x, tail_y-1) # move down

        #otherwise dont move at all
        case _ : return(tail_x, tail_y)


head_x = 0
head_y = 0

tail_x = 0
tail_y = 0

head_positions = [(0,0)]
tail_positions = [(0,0)]

def resolve_move(head_x: int, head_y: int, tail_x: int, tail_y: int):
    head_positions.append((head_x, head_y))
    t_x, t_y = move_tail(head_x, head_y, tail_x, tail_y)
    tail_positions.append((t_x, t_y))
    return t_x,t_y

for move in head_moves:
    match move.split():
        case ["R", count]: 
            for _ in range(int(count)):
                head_x+=1
                tail_x, tail_y = resolve_move(head_x, head_y, tail_x, tail_y)
        case ["L", count]: 
            for _ in range(int(count)):
                head_x-=1
                tail_x, tail_y = resolve_move(head_x, head_y, tail_x, tail_y)
        case ["U", count]:
            for _ in range(int(count)):
                head_y+=1
                tail_x, tail_y = resolve_move(head_x, head_y, tail_x, tail_y)
        case ["D", count]: 
            for _ in range(int(count)):
                head_y-=1
                tail_x, tail_y = resolve_move(head_x, head_y, tail_x, tail_y)
```

```python
x = []
y = []
ids = []
time = []

head_x_positions, head_y_positions = list(map(list, zip(*head_positions)))
tail_x_positions, tail_y_positions = list(map(list, zip(*tail_positions)))

x.extend(head_x_positions)
x.extend(tail_x_positions)
y.extend(head_y_positions)
y.extend(tail_y_positions)
ids.extend(['head' for _ in head_x_positions])
ids.extend(['tail' for _ in tail_x_positions])
time.extend([i for i in range(len(head_x_positions))])
time.extend([i for i in range(len(head_x_positions))])

df = pd.DataFrame({
    "x": x,
    "y": y,
    'id': ids,
    'time': time
})

df
```

```python
class Coordinate:
    def __init__(self, x:int, y:int):
        self.x = x
        self.y = y
    def __str__(self) -> str:
        return f'[{self.x}, {self.y}]'
    def __repr__(self) -> str:
        return self.__str__()

def move_snake(prev: Coordinate, knot: Coordinate):
    diff_x, diff_y = prev.x - knot.x, prev.y - knot.y
    match diff_x, diff_y:
        # diagonal moves
        case(1,2) | (2,1) | (2,2): return Coordinate(knot.x+1, knot.y+1) # move right+up
        case(1,-2) | (2, -1) | (2, -2): return Coordinate(knot.x+1, knot.y-1) # move right+down
        case(-1,2) | (-2,1) | (-2, 2): return Coordinate(knot.x-1, knot.y+1) # move left+up
        case(-1,-2) | (-2,-1) | (-2,-2): return Coordinate(knot.x-1, knot.y-1) # move left+down

        #horizontal moves
        case(2,0): return Coordinate(knot.x+1, knot.y) # move right
        case(-2,0): return Coordinate(knot.x-1, knot.y) # move left

        #vertical moves
        case(0,2): return Coordinate(knot.x, knot.y+1) # move up
        case(0,-2): return Coordinate(knot.x, knot.y-1) # move down

        #otherwise dont move at all
        case _ : return Coordinate(knot.x, knot.y)

knots = [Coordinate(0,0) for i in range(10)]

snake_tail_positions = []

def resolve_snake():
    # print(knots[0])
    for idx, knot in enumerate(knots):
        if idx == 0: 
            pass
        else:
            prev = knots[idx-1]
            #print(f'{idx} : {prev}, {knot}')
            new_position = move_snake(prev, knot)
            #print(new_position)
            knots[idx] = new_position
    snake_tail_positions.append(knots[-1])
    #print('')

for move in head_moves:
    match move.split():
        case ["R", count]: 
            for _ in range(int(count)):
                knots[0].x = knots[0].x + 1
                resolve_snake()
        case ["L", count]: 
            for _ in range(int(count)):
                knots[0].x = knots[0].x - 1
                resolve_snake()
        case ["U", count]:
            for _ in range(int(count)):
                knots[0].y = knots[0].y + 1
                resolve_snake()
        case ["D", count]: 
            for _ in range(int(count)):
                knots[0].y = knots[0].y - 1
                resolve_snake()
```
