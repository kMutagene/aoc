```python
import plotly.express as px
```

```python
tree_grid = []

with open("input.txt","r") as f:
    for line in f:
        tree_grid.append(list(map(lambda s: int(s), list(line.replace("\n","")))))
```

```python
px.imshow(tree_grid, text_auto=True)
```

```python
visible_trees = [
    [0] * len(tree_grid[0]) for _ in range(len(tree_grid))
]

row_count = len(tree_grid)
col_count = len(tree_grid[0])

for row_index in range(len(tree_grid)):
    highest_tree = 0
    for col_index in range(len(tree_grid[0])):
    # visible from the left: traverse rows from left to right
        tree = tree_grid[row_index][col_index] 
        if tree > highest_tree or col_index == 0:
            visible_trees[row_index][col_index] = 1
            highest_tree = tree
    highest_tree = 0
    for col_index in reversed(range(len(tree_grid[0]))):
    # visible from the left: traverse rows from left to right
        tree = tree_grid[row_index][col_index] 
        if tree > highest_tree or col_index == col_count - 1:
            visible_trees[row_index][col_index] = 1
            highest_tree = tree

for col_index in range(len(tree_grid[0])):
    highest_tree = 0
    # visible from the top: traverse cols from top to bottom
    for row_index in range(len(tree_grid)):
        tree = tree_grid[row_index][col_index] 
        if tree > highest_tree or row_index == 0:
            visible_trees[row_index][col_index] = 1
            highest_tree = tree
    highest_tree = 0
    # visible from the bottom: traverse cols from bottom to top
    for row_index in reversed(range(len(tree_grid))):
        tree = tree_grid[row_index][col_index] 
        if tree > highest_tree or row_index == row_count - 1:
            visible_trees[row_index][col_index] = 1
            highest_tree = tree
```

```python
px.imshow(visible_trees, text_auto=True)
```

```python
from itertools import chain

result = sum(chain(*visible_trees))

result
```

```python
scores = []

for row_index in range(len(tree_grid)):
    for col_index in range(len(tree_grid[0])):
        tree = tree_grid[row_index][col_index]
        left_score, right_score, top_score, bottom_score = 0, 0, 0, 0
        l_index = col_index - 1
        r_index = col_index + 1
        t_index = row_index - 1
        b_index = row_index + 1
        l_scenic, r_scenic, t_scenic, b_scenic = True, True, True, True

        while l_scenic and l_index >=0 :
            if tree_grid[row_index][l_index] >= tree:
                l_scenic = False
            left_score += 1
            l_index -= 1

        while r_scenic and r_index < col_count :
            if tree_grid[row_index][r_index] >= tree:
                r_scenic = False
            right_score += 1
            r_index += 1

        while t_scenic and t_index >=0 :
            if tree_grid[t_index][col_index] >= tree:
                t_scenic = False
            top_score += 1
            t_index -= 1

        while b_scenic and b_index < row_count :
            if tree_grid[b_index][col_index] >= tree:
                b_scenic = False
            bottom_score += 1
            b_index += 1

        scores.append(left_score * right_score * top_score * bottom_score)

max(scores)
```
